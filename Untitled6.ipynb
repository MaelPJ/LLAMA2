```python
import torch
import transformers
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-chat-hf",
                                          use_auth_token=True,)

model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-2-7b-chat-hf",
                                             device_map='auto',
                                             torch_dtype=torch.float16,
                                             use_auth_token=True,)
```

```python
from transformers import pipeline

pipe = pipeline("text-generation",
                model=model,
                tokenizer= tokenizer,
                torch_dtype=torch.bfloat16,
                device_map="auto",
                do_sample=True,
                top_k=30,
                num_return_sequences=1,
                eos_token_id=tokenizer.eos_token_id)
```

```python
import json
import textwrap

B_INST, E_INST = "[INST]", "[/INST]"
B_SYS, E_SYS = "<<SYS>>\n", "\n<<SYS>>\n\n"

def get_prompt(instruction, base_prompt):
  SYSTEM_PROMPT = B_SYS + base_prompt + E_SYS
  prompt_template = B_INST + SYSTEM_PROMPT + instruction + E_INST
  return prompt_template

def cut_off_text(text, prompt):
  cutoff_phrase = prompt
  index = text.find(cutoff_phrase)
  if index != -1:
    return text[:index]
  else:
    return text

def remove_substring(string, substring):
  return string.replace(substring, "")

def generate(text, base_prompt):
  prompt = get_prompt(text, base_prompt)
  with torch.autocast('cuda', dtype=torch.float16):
    inputs = tokenizer(prompt, return_tensors="pt").to('cuda')
    outputs = model.generate(**inputs,
                             max_new_tokens=768,
                             eos_token_id=tokenizer.eos_token_id,
                             pad_token_id=tokenizer.eos_token_id,
                             )
    final_outputs = tokenizer.batch_decode(outputs, skip_special_tokens=True)[0]
    final_outputs = cut_off_text(final_outputs, '</s>')
    final_outputs = remove_substring(final_outputs, prompt)

  return final_outputs

def parse_text(text):
 wrapped_text = textwrap.fill(text, width=100)
 return wrapped_text + '\n\n'
```

```python
import os
from google.colab import drive
drive.mount('/content/drive')

topics = ['Golden Retriever']
prompt_beginner = 'Write a 1000 words text in a format readable for a 5 year old about'
prompt_intermediate = 'Write a 1000 words text in a format readable for an average adult about'
prompt_advanced = 'Write a 1000 words scientific article, that could only be understood by an expert about'


def write_text(topic, file_name, text):
  folder_path = os.path.join('/content/drive/My Drive/Generated_Texts', topic)
  os.makedirs(folder_path, exist_ok=True)

  file_path = os.path.join(folder_path, file_name)

  with open(file_path, 'w') as file:
      file.write(text)

for topic in topics:

   write_text(topic, 'beginner.txt', parse_text(generate(topic,prompt_beginner)))
   write_text(topic, 'intermediate.txt', parse_text(generate(topic,prompt_intermediate)))
   write_text(topic, 'advanced.txt', parse_text(generate(topic,prompt_advanced)))

```
